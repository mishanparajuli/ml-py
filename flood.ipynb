```python
target= df.Flood
target.head(10)
```

```python
dic={'0.0': 'Flood'}
target.rename(column=dic, inplace=True)
```

```python
x = df.iloc[0:20545,4:13].values
```

```python
print(x.shape)
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, target, test_size=0.2, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
flood_model= MultinomialNB()
```

```python
flood_model.fit(x_train, y_train)
```

```python
y_pred = flood_model.predict(x_test)
```

```python
print(y_pred)
print(y_pred[-1])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
score = accuracy_score(y_test, y_pred)
print(f" Flood Detection Accuracy: {round(score*100,2)}%")
```
